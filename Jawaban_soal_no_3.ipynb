```python
import socket

HOST = "127.0.0.1"
PORT = 5050

sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# =============== TIMEOUT CONNECT ===============
sock.settimeout(3)  # timeout 3 detik saat connect

print("Mencoba connect ke server...")
try:
    sock.connect((HOST, PORT))
    print("Berhasil connect!")
except socket.timeout:
    print("Koneksi timeout! (saat connect)")
    sock.close()
    raise SystemExit
except Exception as e:
    print("Gagal connect:", e)
    sock.close()
    raise SystemExit


# =============== TIMEOUT READ ===============
sock.settimeout(2)  # timeout 2 detik saat membaca data

print("Menunggu data dari server...")

try:
    data = sock.recv(1024).decode()
    print("Dari server:", data)
except socket.timeout:
    print("Koneksi timeout! (saat membaca data)")
except Exception as e:
    print("Error:", e)

sock.close()
```
